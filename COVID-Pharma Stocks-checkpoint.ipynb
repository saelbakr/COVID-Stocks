```python
# Import COVID confirmed cases data
data_COVID = "output/who_covid_19_sit_rep_time_series.csv"
data_COVID_df = pd.read_csv(data_COVID)
data_COVID_df.head()
```

```python
data_non_US = data_COVID_df

data_non_US.head()
```

```python
data_non_US.columns
```

```python
#Extract the confirmed cases for Non US countries from Jan to March
data_non_us_months = data_non_US[['1/21/2020', '1/22/2020', '1/23/2020', '1/24/2020', '1/25/2020', '1/26/2020', '1/27/2020', '1/28/2020', '1/29/2020', '1/30/2020', '1/31/2020', '2/1/2020', '2/2/2020', '2/3/2020', '2/4/2020', '2/5/2020', '2/6/2020', '2/7/2020', '2/8/2020', '2/9/2020', '2/10/2020', '2/11/2020', '2/12/2020', '2/13/2020', '2/14/2020', '2/15/2020', '2/16/2020', '2/17/2020', '2/18/2020', '2/19/2020', '2/20/2020', '2/21/2020', '2/22/2020', '2/23/2020', '2/24/2020', '2/25/2020', '2/26/2020', '2/27/2020', '2/28/2020', '2/29/2020', '3/1/2020', '3/2/2020', '3/3/2020', '3/4/2020', '3/5/2020']]
data_non_us_months
```

```python
data_non_us_sum = pd.DataFrame({"Total":data_non_us_months.sum()})
```

```python

data_non_us_sum["Date1"] = data_non_us_sum.index

data_non_us_sum["Date1"] = pd.to_datetime(data_non_us_sum["Date1"])
```

```python
data_non_us_sum.dtypes
```

```python
data_non_us_sum 
```

```python
#data_close.rename(columns = {"Date":"Date1"})
data_close["Date1"] = data_close["Date"]
data_close.dtypes
```

```python
combined = pd.merge(data_non_us_sum ,data_close, how = "inner", on ="Date1")
combined
```

```python
combined.rename(columns = {"Total_x":"COVID_Cases", "Total_y":"Stock_Prices"})
```

```python
combined.columns
```

```python
# Scatter plot for top pharma stocks and COVID cases

x = combined["Total_x"]
y = combined["Total_y"]
plt.scatter(x,y)
plt.xlabel('Confirmed COVID Cases')
plt.ylabel('Pharma Stock Prices')
plt.show()

correlation = st.pearsonr(x,y)
print(f"The correlation between COVID cases and Pharma Stocks is {round(correlation[0],2)}")
```
